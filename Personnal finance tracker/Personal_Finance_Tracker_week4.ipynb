```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 4: Loop-Powered Edition")
print()

# User setup
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# MULTI-EXPENSE ENTRY
expenses = []          # Store all expense amounts
descriptions = []      # Store what each expense was for
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == "done":
        break

    try:
        expense = float(expense_input)

        if expense <= 0:
            print("Expense must be a positive number!")
            continue

        description = input("Enter description: ").strip()
        while description == "":
            print("Description cannot be empty!")
            description = input("Enter description: ").strip()

        expenses.append(expense)
        descriptions.append(description)
        total_spent += expense

        budget_percentage = (total_spent / monthly_budget) * 100

        print(f"Added: ${expense:.2f} - {description}")
        print(f"Running total: ${total_spent:.2f}")
        print(f"Budget used: {budget_percentage:.1f}%")

    except ValueError:
        print("Please enter a valid expense amount!")

print(f"\nExpense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")


# PART 2: BASIC EXPENSE ANALYSIS
if len(expenses) > 0:
    print("\nExpense Analysis:")

    print(f"Total expenses: {len(expenses)}")
    print(f"Total spent: ${total_spent:.2f}")
    print(f"Average expense: ${total_spent / len(expenses):.2f}")
    print(f"Budget used: {(total_spent / monthly_budget) * 100:.1f}%")
    print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    # Find highest and lowest expenses
    highest_expense = expenses[0]
    lowest_expense = expenses[0]
    highest_desc = descriptions[0]
    lowest_desc = descriptions[0]

    for i in range(len(expenses)):
        if expenses[i] > highest_expense:
            highest_expense = expenses[i]
            highest_desc = descriptions[i]
        if expenses[i] < lowest_expense:
            lowest_expense = expenses[i]
            lowest_desc = descriptions[i]

    print("\nExpense Range:")
    print(f"Highest expense: ${highest_expense:.2f} - {highest_desc}")
    print(f"Lowest expense: ${lowest_expense:.2f} - {lowest_desc}")

    # Categorize expenses
    large_count = 0   # > 20% of budget
    medium_count = 0  # 5-20% of budget
    small_count = 0   # < 5% of budget

    for expense in expenses:
        percent_of_budget = (expense / monthly_budget) * 100

        if percent_of_budget > 20:
            large_count += 1
        elif percent_of_budget >= 5:
            medium_count += 1
        else:
            small_count += 1

    print("\nExpense Categories:")
    print(f"Large expenses (>20% budget): {large_count}")
    print(f"Medium expenses (5-20% budget): {medium_count}")
    print(f"Small expenses (<5% budget): {small_count}")

    # Smart feedback
    print("\nSmart Feedback:")
    budget_used = (total_spent / monthly_budget) * 100

    if budget_used >= 100:
        print("You have exceeded your monthly budget. Reduce non-essential spending immediately.")
    elif budget_used >= 80:
        print("You are close to your budget limit. Be careful with the rest of your spending.")
    elif budget_used >= 50:
        print("You have used over half your budget. Keep monitoring your spending.")
    else:
        print("Your spending is under control so far.")

    if large_count >= 2:
        print("You have several large expenses. Review whether all of them were necessary.")
    elif large_count == 1:
        print("You had one major expense this period. Plan around it carefully.")
    else:
        print("Most of your expenses are relatively small, which is easier to manage.")

else:
    print("\nNo expenses entered - try adding some transactions!")


# PART 3: MULTIPLE SAVINGS GOALS TRACKER 
print("\n=== Savings Goals Tracker ===")

goal_names = []
goal_targets = []
goal_current = []

num_goals = 0
while num_goals < 1 or num_goals > 5:
    try:
        num_goals = int(input("How many savings goals? (1-5): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 5:
            print("Maximum 5 goals for this tracker!")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nEntering {num_goals} savings goals:")

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    goal_name = input("Goal name: ").strip()
    while goal_name == "":
        print("Goal name cannot be empty!")
        goal_name = input("Goal name: ").strip()

    target_amount = -1
    while target_amount <= 0:
        try:
            target_amount = float(input("Target amount: $"))
            if target_amount <= 0:
                print("Target amount must be greater than $0!")
        except ValueError:
            print("Please enter a valid number!")

    current_amount = -1
    while current_amount < 0 or current_amount > target_amount:
        try:
            current_amount = float(input("Current saved amount: $"))
            if current_amount < 0:
                print("Current amount cannot be negative!")
            elif current_amount > target_amount:
                print("Current amount cannot be more than the target!")
        except ValueError:
            print("Please enter a valid number!")

    goal_names.append(goal_name)
    goal_targets.append(target_amount)
    goal_current.append(current_amount)

print(f"\nSaved {len(goal_names)} goals!")


# GOAL ANALYSIS AND PROGRESS TRACKING 
if len(goal_names) > 0:
    print("\nGoals Progress Analysis:")

    total_target = 0
    total_current = 0

    lowest_progress = None
    attention_goal = ""

    for i in range(len(goal_names)):
        progress = (goal_current[i] / goal_targets[i]) * 100
        remaining = goal_targets[i] - goal_current[i]
        months_to_completion = remaining / 100

        print(f"\n{goal_names[i]}")
        print(f"Saved: ${goal_current[i]:.2f} / ${goal_targets[i]:.2f}")
        print(f"Progress: {progress:.1f}%")
        print(f"Remaining: ${remaining:.2f}")
        print(f"Months to completion at $100/month: {months_to_completion:.1f}")

        total_target += goal_targets[i]
        total_current += goal_current[i]

        if lowest_progress is None or progress < lowest_progress:
            lowest_progress = progress
            attention_goal = goal_names[i]

    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\nOverall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")

    print("\nGoal Needing Most Attention:")
    print(f"{attention_goal} has the lowest progress at {lowest_progress:.1f}%.")

    if lowest_progress < 25:
        print("This goal is still at an early stage. Consider giving it a dedicated monthly amount.")
    elif lowest_progress < 75:
        print("This goal is underway, but it still needs consistent progress.")
    else:
        print("This goal is close to completion. Keep going.")

else:
    print("No goals to analyze!")


#  PART 4: SIMPLE MENU SYSTEM 
print("\n=== Interactive Finance Menu ===")

running = True

while running:
    print("\n" + "-" * 30)
    print(f"{name}'s Finance Tracker")
    if len(expenses) > 0:
        print(f"Current: {len(expenses)} expenses, ${sum(expenses):.2f} spent")
    else:
        print("Current: 0 expenses, $0.00 spent")
    print("-" * 30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    if choice == "1":
        print("\n--- Expense Summary ---")

        if len(expenses) == 0:
            print("No expenses recorded.")
        else:
            for i in range(len(expenses)):
                print(f"{i+1}. ${expenses[i]:.2f} - {descriptions[i]}")
            print(f"Total: ${sum(expenses):.2f}")

    elif choice == "2":
        print("\n--- Goals Progress ---")

        if len(goal_names) == 0:
            print("No goals recorded.")
        else:
            for i in range(len(goal_names)):
                progress = (goal_current[i] / goal_targets[i]) * 100
                print(f"{goal_names[i]}: ${goal_current[i]:.2f}/${goal_targets[i]:.2f} ({progress:.1f}% complete)")

    elif choice == "3":
        print("\n--- Add New Expense ---")

        while True:
            new_expense_input = input("Expense amount: $")
            try:
                new_expense = float(new_expense_input)

                if new_expense <= 0:
                    print("Expense must be a positive number!")
                    continue

                new_description = input("Enter description: ").strip()
                while new_description == "":
                    print("Description cannot be empty!")
                    new_description = input("Enter description: ").strip()

                expenses.append(new_expense)
                descriptions.append(new_description)
                total_spent += new_expense

                print(f"Added: ${new_expense:.2f} - {new_description}")
                print(f"New total spent: ${total_spent:.2f}")
                break

            except ValueError:
                print("Please enter a valid expense amount!")

    elif choice == "4":
        print("\n--- Budget Check ---")

        spent = sum(expenses)
        remaining = monthly_budget - spent
        percent_used = (spent / monthly_budget) * 100
        daily_budget_left = remaining / 30

        print(f"Monthly budget: ${monthly_budget:.2f}")
        print(f"Spent so far: ${spent:.2f}")
        print(f"Budget used: {percent_used:.1f}%")
        print(f"Remaining amount: ${remaining:.2f}")
        print(f"Daily budget left (30-day month): ${daily_budget_left:.2f}")

        if percent_used > 100:
            print("Advice: You are over budget. Cut spending where possible.")
        elif percent_used > 80:
            print("Advice: You are nearing your budget limit.")
        elif percent_used > 50:
            print("Advice: You are in the middle range of your budget. Stay aware.")
        else:
            print("Advice: Your budget is in a healthy position.")

    elif choice == "5":
        running = False
        print(f"\nGoodbye {name}!")
        print(f"Final summary: {len(expenses)} expenses, ${sum(expenses) if expenses else 0:.2f} total")
        print("Your finance tracker is growing stronger each week!")

    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```

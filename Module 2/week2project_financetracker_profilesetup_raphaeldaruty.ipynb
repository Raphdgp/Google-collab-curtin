Goal: Create a simple program that takes input, setups user profile and creates a custom welcome message for a finance tracker.

```python
#Stores user data in variables
username = input("Type your username here: ")
theme = input("Type preferred theme color here: ")
primary_currency = input("Type your primary currency here: ")
saving_goal = input("Type your main reason for saving: ")

#Create the message
welcome_message = f"Greetings, {username}! Your preferred theme color is {theme}, your primary currency is {primary_currency}, and your main reason for saving is {saving_goal}."

#print the welcome message
print(welcome_message)

```

```python
expense = float(input("How much was the expense? "))

if expense > 100:
    print("Major expense")
elif expense > 50:
    print("Moderate expense")
elif expense > 10:
    print("Minor expense")
else:
    print("Small purchase")
```

Part 2: Financial Decision Logic

```python
print("=== Budget Limit Checker ===")

# Get user's budget and expense
monthly_budget = float(input("Enter your monthly budget: $"))
expense = float(input("Enter expense amount: $"))

if expense > 0.5 * monthly_budget:
    print("Major purchase, think carefully")
elif expense > 0.25 * monthly_budget:
    print("Significant expense")
elif expense > 0.1 * monthly_budget:
    print("Moderate expense")
else:
    print("Minor expense, within budget")
```

```python
print("=== Savings Goal Tracker ===")

goal_amount = float(input("Savings goal: $"))
current_savings = float(input("Current savings: $"))
savings_percentage = (current_savings / goal_amount) * 100

if savings_percentage >= 100:
    print("Goal achieved! Consider a new goal.")
elif savings_percentage >= 75:
    print("Almost there! You're doing great.")
elif savings_percentage >= 50:
    print("Halfway there - keep it up!")
elif savings_percentage >= 25:
    print("Good progress, stay consistent")
else:
    print("Just getting started - you've got this!")
```

Part 3: Smart Purchase Advisor

```python
print("=== Smart Purchase Advisor ===")

# Get financial information
available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

# Calculate affordability percentage
affordability_ratio = purchase_price / available_money

if affordability_ratio > 1:
    print("You cannot afford this purchase right now.")
elif affordability_ratio > 0.5:
    if purchase_price > 0.5 * monthly_income:
        print("It's a large part of your available money and of your monthly income. Only consider if it's a critical purchase.")
    elif purchase_price > 0.25 * monthly_income:
        print("It's a large part of your available money, and a significant part of your monthly income. Be very careful")
    elif purchase_price > 0.1 * monthly_income:
        print("It's a large part of your available money and small part of your monthly income. Make sure you will have enough to finish the month.")
    else:
        print("It's a large part of your available money but a very small part of your monthly income. It's probably fine")
elif affordability_ratio > 0.25:
    if purchase_price > 0.5 * monthly_income:
        print("It's a moderate part of your available money and a large part of your monthly income. Only consider if it's a critical purchase.")
    elif purchase_price > 0.25 * monthly_income:
        print("It's a moderate part of your available money, and a significant part of your monthly income. Be very careful")
    elif purchase_price > 0.1 * monthly_income:
        print("It's a moderate part of your available money and small part of your monthly income. Make sure you will have enough to finish the month.")
    else:
        print("It's a moderate part of your available money but a very small part of your monthly income. It's probably fine")
elif affordability_ratio > 0.1:
    if purchase_price > 0.5 * monthly_income:
        print("It's a small part of your available money but a large part of your monthly income. Don't take out of your savings if it's hard to replace.")
    elif purchase_price > 0.25 * monthly_income:
        print("It's a small part of your available money but a significant part of your monthly income. Are you sure you need this?")
    elif purchase_price > 0.1 * monthly_income:
        print("It's a small part of your available money and a small part of your monthly income. Go ahead!")
    else:
        print("It's a small part of your available money and a very small part of your monthly income. Just buy it!")
else:
    if purchase_price > 0.5 * monthly_income:
        print("It's a very small part of your available money but a large part of your monthly income. Don't take out of your savings if it's hard to replace.")
    elif purchase_price > 0.25 * monthly_income:
        print("It's a very small part of your available money but a significant part of your monthly income. Are you sure you need this?")
    elif purchase_price > 0.1 * monthly_income:
        print("It's a very small part of your available money and a small part of your monthly income. Go ahead!")
    else:
        print("It's a very small part of your available money and a very small part of your monthly income. Buy 2 of it if you want!")
```
